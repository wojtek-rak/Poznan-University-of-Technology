```python
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt


from sympy.interactive import printing
printing.init_printing(use_latex=True)


import sympy as sym
from sympy import *

x, z = symbols("x z")

ran = np.arange(-5,5,0.1)

eq = - x**3 + 3*x**2 + 10*x - 24
eq

y = []

for l in ran:
    y.append(eq.evalf(subs={x:l}))
    


plt.plot(ran, y)
plt.show()

solve(eq, x)

```

```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
import sympy as sym
# Definicje zmiennych
x, y = sym.symbols("x y")
#eq1 = x**2 + 3*y = 10
#eq2 = 4*x -y**2 = -2
solve((x**2 + 3*y - 10, 4*x -y**2 + 2), (x,y))
```

```python
for k, v in solve((x**2 + 3*y - 10, 4*x -y**2 + 2), (x,y)):
    print( v.evalf())
```
