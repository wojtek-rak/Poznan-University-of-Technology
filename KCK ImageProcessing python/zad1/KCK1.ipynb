```python
#!/ usr/bin/env python 
#coding: utf8

%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd 

def main(): 
    data1 = pd.read_csv("2cel-rs.csv") 
    data2 = pd.read_csv("2cel.csv")
    data3 = pd.read_csv("cel-rs.csv") 
    data4 = pd.read_csv("cel.csv") 
    data5 = pd.read_csv("rsel.csv") 

    for d in [data1, data2, data3, data4, data5]:
        l = d.drop(columns = ['generation', 'effort'])
        d['avg'] = l.mean(axis=1)

    plt.rcParams["font.size"] = "15"    
    
    plt.figure(figsize=(6.7,5))
    plt.margins(0.05, 0.05)
    plt.plot (data1['effort'],data1['avg'],label='2-Coev-RS',color='r',linewidth=1.7) 
    plt.plot (data1['effort'],data2['avg'],label='2-Coev',color='m',linewidth=1.7) 
    plt.plot (data1['effort'],data3['avg'],label='1-Coev-RS',color='g',linewidth=1.7) 
    plt.plot (data1['effort'],data4['avg'],label='1-Coev',color='k',linewidth=1.7) 
    plt.plot (data1['effort'],data5['avg'],label='1-Evol-RS',color='b',linewidth=1.7) 
    
    plt.ylabel("Odsetek wygranych gier")
    plt.xlabel("Rozegranych gier")
    
    plt.legend(
        loc="lower right",
        prop={'size': 13},
        edgecolor='black',
        facecolor='white',
        framealpha=1.0,
        fancybox=0)
    
    
    
    plt.show()
    
    
    
    #plt.savefig('myplot.pdf')
    plt.close ()
    
if __name__ == '__main__': 
    main()
```
